```python
# Import packages
import pandas as pd
```

```python
# Import ISOT

ISOT_Fake_dataset = pd.read_csv("../datasets/ISOT/Fake.csv")
ISOT_True_dataset = pd.read_csv("../datasets/ISOT/True.csv")

ISOT_Fake_dataset.head()
```

```python
ISOT_True_dataset.head()
```

```python
# Add column for label and assign label values

ISOT_Fake_dataset['label'] = 1
ISOT_Fake_dataset
```

```python
# Add column for label and assign label values

ISOT_True_dataset['label'] = 0
print(ISOT_True_dataset)
```

```python
# Combine ISOT datasets
ISOT_combined_dataset = pd.concat([ISOT_Fake_dataset, ISOT_True_dataset], ignore_index=True)

# Re-arrange columns
ISOT_combined_dataset = ISOT_combined_dataset[['text', 'label', 'title', 'subject', 'date']]
ISOT_combined_dataset
```

```python
# Subjects to filter by
print("ISOT Subjects:",ISOT_combined_dataset['subject'].value_counts())
```

```python
# Filter for subject

# Science
ISOT_science = ISOT_combined_dataset[ISOT_combined_dataset['text'].str.contains('climate change')]
ISOT_combined_dataset = pd.concat([ISOT_combined_dataset, ISOT_science]).drop_duplicates(keep=False)

# Politics
ISOT_politics = ISOT_combined_dataset[ISOT_combined_dataset['subject'].isin(['politicsNews', 'politics', 'left-news'])]

ISOT_combined_dataset = pd.concat([ISOT_combined_dataset, ISOT_politics]).drop_duplicates(keep=False)


# Social Media
# Remove previous filtered rows and make remaining ones for social media
ISOT_social = ISOT_combined_dataset
ISOT_social
```

```python
# Save split up dataframes as raw datasets
# ISOT Science
ISOT_science.to_csv("../datasets/science/isot_science.csv")

# ISOT Politics
ISOT_politics.to_csv("../datasets/politics/isot_politics.csv")

# ISOT Social
ISOT_social.to_csv("../datasets/social-media/isot_social.csv")
```

```python
# Add metadata
ISOT_science = ISOT_science[['text', 'label']].assign(metadata=ISOT_science.iloc[:,2:].agg(dict,1))

ISOT_politics = ISOT_politics[['text', 'label']].assign(metadata=ISOT_politics.iloc[:,2:].agg(dict,1))

ISOT_social = ISOT_social[['text', 'label']].assign(metadata=ISOT_social.iloc[:,2:].agg(dict,1))
ISOT_social
```

```python
# Add column for classification category
ISOT_science['classification'] = 3

ISOT_science = ISOT_science[['text', 'label', 'classification', 'metadata']]

ISOT_science.head()
```

```python
# Add column for classification category
ISOT_politics['classification'] = 2

ISOT_politics = ISOT_politics[['text', 'label', 'classification', 'metadata']]

ISOT_politics.head()
```

```python
# Add column for classification category
ISOT_social['classification'] = 4

ISOT_social = ISOT_social[['text', 'label', 'classification', 'metadata']]

ISOT_social.head()
```

```python
# Save as Pyarrow Feather format

# ISOT Science
ISOT_science.to_feather("../preprocessed_datasets/science/isot_science_part.feather")

# ISOT Politics
ISOT_politics.to_feather("../preprocessed_datasets/politics/isot_politics_part.feather")

# ISOT Social
ISOT_social.to_feather("../preprocessed_datasets/social-media/isot_social_part.feather")
```
